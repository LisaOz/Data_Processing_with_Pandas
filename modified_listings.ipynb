```python
import pandas as pd
df = pd.read_csv('listings.csv')
print(df.head(10))
print(df.tail(10))
```

```python
df.columns
```

```python
df.describe()
```

```python
# Convert all columns to 'object' type to allow string replacements
df = df.astype('object')

# Replace NaN values with 'M'
df.fillna('M', inplace=True)

# Print the modified DataFrame
print(df)
```

```python
# display only the 'host_id', 'price' and 'availability' columns
selected_columns = df[['host_id', 'price', 'minimum_nights', 'availability_365']]
print("\nSelected Columns:")
print(selected_columns)
```

```python
# rows where the price is > 90
filtered_df = df[df['price'] > 90]

# Selected columns: 'host_id', 'price', and 'availability_365'
selected_columns = filtered_df[['host_id', 'price', 'availability_365']]

# Display the result
print("\nSelected Columns:")
print(selected_columns)
```

```python
# Sort the filtered DataFrame by 'price' in ascending order
sorted_df = selected_columns.sort_values(by='price', ascending=True)

# Display the result
print("\nSelected Columns Sorted by Price (Ascending):")
print(sorted_df)
```

```python
# Save the modified DataFrame to a CSV file
df.to_csv('modified_listings.csv', index=False)

print("Modified DataFrame saved as 'modified_listings.csv'.")
```

```python
# Zip the CSV file
import zipfile
with zipfile.ZipFile('modified_listings.zip', 'w') as zipf:
    zipf.write('modified_listings.csv')
```
